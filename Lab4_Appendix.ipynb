```python
import sys
from collections import defaultdict, Counter, deque

SEED_NUM = 8  
STUDENT_MAJOR = "BSECE"
FAVORITE_ARTIST = "ERASERHEADS"
LOCAL_HAZARD = "EARTHQUAKE"
CONTROL_NUM = max(3, SEED_NUM)

print(f"SYSTEM INITIALIZED | CONTROL_NUM: {CONTROL_NUM}")
```

```python
# 1. Initialize Stream
stream = [
    ("TUP Manila", "BSECE"),
    ("TUP Taguig", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", f"BSME_v{CONTROL_NUM}"),
    ("TUP Manila", "BSECE"),
    ("TUP Visayas", STUDENT_MAJOR),
    ("TUP Taguig", "BSECE"),
    ("TUP Manila", "BSECE")
]

# Data Expansion: Add (CONTROL_NUM + 3) = 11 custom tuples
for i in range(11):
    stream.append(("TUP Manila", f"PROGRAM_{i}"))

# Phase 1: Aggregation using defaultdict
campus_map = defaultdict(list)
for campus, program in stream:
    campus_map[campus].append(program)

# Phase 2: Frequency Analysis
print("--- Exercise 1: Manila Campus Analysis ---")
manila_counts = Counter(campus_map["TUP Manila"])
top_program = manila_counts.most_common(1) # Extract top requested program

print(f"Top Requested Program in Manila: {top_program}")
```

```python
# Initialize Sets
festival = {"Ben&Ben", "SB19", "Bini", "Eraserheads", FAVORITE_ARTIST, "Zild", f"Indie Artist {CONTROL_NUM}"}
user_a = {"Ben&Ben", "Bini", "Maki", "Dionela", FAVORITE_ARTIST}
user_b = {"SB19", "Eraserheads", "Zild", f"Indie Artist {CONTROL_NUM}", "Parokya ni Edgar"}

# Data Expansion: Add CONTROL_NUM (8) new artists
for i in range(CONTROL_NUM):
    festival.add(f"Opm_Rising_{i}")

# Jaccard Algorithm
def get_jaccard(user_set, fest_set):
    intersection = user_set.intersection(fest_set)
    union = user_set.union(fest_set)
    return (len(intersection) / len(union)) * 100

print("--- Exercise 2: Similarity Results ---")
print(f"User A Similarity: {get_jaccard(user_a, festival):.2f}%")
print(f"User B Similarity: {get_jaccard(user_b, festival):.2f}%")

# Set Difference for Dealbreakers
dealbreakers = user_a.difference(festival)
print(f"User A Dealbreakers (Artists not in lineup): {dealbreakers}")
```

```python
window = deque(maxlen=5)

burst = [
    (CONTROL_NUM, "WEATHER"),
    (CONTROL_NUM + 1, "TRAFFIC"),
    (CONTROL_NUM + 2, "WEATHER"),
    (CONTROL_NUM + 3, "WEATHER"),
    (CONTROL_NUM + 4, LOCAL_HAZARD), # Using EARTHQUAKE here
    (CONTROL_NUM + 5, "WEATHER"),
    (CONTROL_NUM + 6, "TRAFFIC")
]

# Data Expansion: Add (CONTROL_NUM + 2) = 10 custom alerts
current_ts = burst[-1][0]
for i in range(1, 11):
    category = "TRAFFIC" if i % 2 == 0 else LOCAL_HAZARD
    burst.append((current_ts + i, category))

# Algorithm: Sliding Window with Spam Rule
print("--- Exercise 3: Filter Log ---")
for alert in burst:
    in_window_categories = [item[1] for item in window]
    counts = Counter(in_window_categories)
    
    # Spam Rule: Reject if category exists 2 or more times
    if counts[alert[1]] < 2:
        window.append(alert)
    else:
        print(f"Filtered (Spam Detected): {alert}")

# Final Result
ts_sum = sum(alert[0] for alert in window)
print(f"\nFinal Window State: {list(window)}")
print(f"Sum of Timestamps: {ts_sum}")
```
